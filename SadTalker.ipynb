```python
import pandas as pd

avatares_genero = {
    "avatar0.jpg":"fem",
    "avatar1.jpg":"fem",
    "avatar2.jpg":"masc",
    "avatar3.jpg":"masc",
    "avatar4.jpg":"masc",
    "avatar5.jpg":"masc",
    "avatar6.jpg":"masc",
    "avatar7.jpg":"fem",
    "avatar8.jpg":"fem"
}
path_reference = '/content/drive/MyDrive/TuChisteDiario/videoReferencia.mp4'
df_chistes = pd.read_csv(path + '/chistes_con_metadatos.csv')
```

```python
# Chistes a crear
df_chistes_crear = df_chistes[(~df_chistes['video_creado'])&(df_chistes['audio_creado'])]

for index, row in df_chistes_crear.iterrows():
  chiste = f'chiste_{index}'
  path_chiste = path + '/audios_chistes/' + chiste
  #numero del chiste
  numero_chiste = int(chiste.replace('chiste_',''))
  #numero avatar
  numero_avatar = numero_chiste % len(avatares_genero.keys())
  #imagen avatar
  avatar = 'avatar' + str(numero_avatar) + '.jpg'
  img = path + '/avatares/avatar' + str(numero_avatar) + '.jpg'
  #audio
  audio_masc = path + '/audios_chistes/' + chiste + '/audio_masc'
  audio_fem = path + '/audios_chistes/' + chiste + '/audio_fem'
  audio = audio_masc if avatares_genero[avatar] == 'masc' else audio_fem
  #nombre carpeta
  nombre_carpeta = path + '/videos/' + chiste

  print(f'------- Creando video de {chiste} ---------')

  !python3.8 inference.py --driven_audio {audio} \
            --source_image {img} \
            --result_dir {nombre_carpeta} \
            --still --preprocess full --enhancer gfpgan

  #marcamos chiste como creado
  df_chistes.loc[index,'video_creado'] = True
  df_chistes.to_csv('chistes_con_metadatos.csv',index=False)
```
